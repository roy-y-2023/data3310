```python
ei_stats_filename = "EI-Stats-Review-ALL-data.xlsx"
load_remote_file_if_not_exists(ei_stats_filename, "https://github.com/roy-y-2023/data3310/raw/refs/heads/main/data/EI-Stats-Review-ALL-data.xlsx")

def load_ei_stats_sheet(sheet_name: str, first_row: int, last_row: int) -> pd.DataFrame:
  if first_row >= last_row:
    print("first row cannot be >= last_row")
    return None

  df = (pd.read_excel(
      ei_stats_filename,
      nrows=last_row - first_row + 1,
      skiprows=first_row - 1,
      sheet_name=sheet_name,
    )
    .dropna(how="all")
    .transpose()
    .dropna(how="all")
  )
  df.columns = df.iloc[0]
  df = df.iloc[1:]
  df = df.reset_index()
  return df

gas_prod_df = load_ei_stats_sheet("Gas Production - bcf", 3, 78)
gas_prod_df.columns.values[0] = "Year"
gas_prod_df
```

Dataset: [Energy consumption by source and country](https://ourworldindata.org/grapher/energy-consumption-by-source-and-country)

As its name suggets, it contains different countries for different years for several energy sources (e.g. Solar, Wind, Hydropower, Nuclear, Gas, Coal, Oil). While it contains many data, it doesn't include every country (and some of them on there is blank), so the challenges have been to map the missing ones to a default value like the continent they're on.

```python
# For Visualization 3
energy_consumption_by_source_and_country_filename = "energy-consumption-by-source-and-country.csv"
load_remote_file_if_not_exists(energy_consumption_by_source_and_country_filename, "https://raw.githubusercontent.com/roy-y-2023/data3310/refs/heads/main/data/energy-consumption-by-source-and-country.csv")
energy_consumption_by_source_and_country_df = pd.read_csv(energy_consumption_by_source_and_country_filename)
energy_consumption_by_source_and_country_df.head()
```

```python
CLEAN_ENERGY_SOURCE_NAMES = ["Other renewables", "Biofuels", "Solar", "Wind", "Hydropower", "Nuclear"]
energy_consumption_by_source_and_country_df["Total_Conventional_Energy"] = energy_consumption_by_source_and_country_df["Gas"] + energy_consumption_by_source_and_country_df["Coal"] + energy_consumption_by_source_and_country_df["Oil"]
energy_consumption_by_source_and_country_df["Total_Clean_Energy"] = energy_consumption_by_source_and_country_df["Other renewables"] + energy_consumption_by_source_and_country_df["Biofuels"] + energy_consumption_by_source_and_country_df["Solar"] + energy_consumption_by_source_and_country_df["Wind"] + energy_consumption_by_source_and_country_df["Hydropower"] + energy_consumption_by_source_and_country_df["Nuclear"]
energy_consumption_by_source_and_country_df["Total"] = energy_consumption_by_source_and_country_df["Total_Conventional_Energy"] + energy_consumption_by_source_and_country_df["Total_Clean_Energy"]
energy_consumption_by_source_and_country_df["Clean_Energy_Share"] = energy_consumption_by_source_and_country_df["Total_Clean_Energy"] / energy_consumption_by_source_and_country_df["Total"] * 100
energy_consumption_by_source_and_country_df_2024 = energy_consumption_by_source_and_country_df[energy_consumption_by_source_and_country_df["Year"] == 2024]

energy_consumption_by_source_and_country_df_valid = energy_consumption_by_source_and_country_df_2024.dropna(subset=["Clean_Energy_Share"])
code_to_share = energy_consumption_by_source_and_country_df_valid.set_index("Code")["Clean_Energy_Share"].to_dict()
code_to_entity = energy_consumption_by_source_and_country_df_valid.set_index("Code")["Entity"].to_dict()
entity_to_share = energy_consumption_by_source_and_country_df.set_index("Entity")["Clean_Energy_Share"].to_dict()

def get_continent_share(alpha_2: str) -> float | None:
    try:
        continent_code = pycountry_convert.country_alpha2_to_continent_code(alpha_2)
        continent_name = pycountry_convert.convert_continent_code_to_continent_name(continent_code)
        return entity_to_share.get(continent_name) if continent_name else None
    except Exception:
        return 0

rows = []
for country in pycountry.countries:
    alpha_3 = country.alpha_3
    alpha_2 = country.alpha_2

    if alpha_3 in code_to_share:
        entity = code_to_entity[alpha_3]
        share = code_to_share[alpha_3]
    else:
        entity = country.name
        share = get_continent_share(alpha_2)

    rows.append({"Entity": entity, "Code_Num": int(country.numeric), "Clean_Energy_Share": share})

all_countries_energy_consumption_by_source_df = pd.DataFrame(rows)
all_countries_energy_consumption_by_source_df = all_countries_energy_consumption_by_source_df[all_countries_energy_consumption_by_source_df["Entity"] != "Antarctica"]
all_countries_energy_consumption_by_source_df.head()
```

```python
countries = ['United States', 'Uruguay', 'Nigeria', 'China']

def country_plot(country):
    country_df = df[df['Entity'] == country]

    withE = alt.Chart(country_df).mark_line().encode(
        x=alt.X('Year:O', title='Year'),
        y=alt.Y('With electricity:Q', title='Population'),
    )

    withoutE = alt.Chart(country_df).mark_line(strokeDash=[5,5]).encode(
        x='Year:O',
        y='Without electricity:Q',
    )

    return (withE + withoutE).properties(
        title=country,
        width=300,
        height=200
    )


c1 = country_plot(countries[0])
c2 = country_plot(countries[1])
c3 = country_plot(countries[2])
c4 = country_plot(countries[3])

(c1 | c2) & (c3 | c4)
```

```python
continent_entity_names = ["Africa", ]
world_energy_consumption_by_source_df = energy_consumption_by_source_and_country_df[energy_consumption_by_source_and_country_df["Entity"] == "World"]
world_energy_consumption_by_source_df
```

```python
world_energy_consumption_by_source_and_year_df = world_energy_consumption_by_source_df.melt(
    id_vars=['Year'],
    value_vars=CLEAN_ENERGY_SOURCE_NAMES,
    var_name='Clean Energy Source',
    value_name='Energy Consumption'
)
world_energy_consumption_by_source_and_year_df
```

```python
world_energy_consumption_by_source_and_year_df = world_energy_consumption_by_source_and_year_df.sort_values(['Clean Energy Source', 'Year'])
world_energy_consumption_by_source_and_year_df["Amount Chnage"] = world_energy_consumption_by_source_and_year_df.groupby("Clean Energy Source")["Energy Consumption"].diff()
world_energy_consumption_by_source_and_year_df["Percent Change"] = world_energy_consumption_by_source_and_year_df.groupby("Clean Energy Source")["Energy Consumption"].pct_change() * 100
# world_energy_consumption_by_source_and_year_df = world_energy_consumption_by_source_and_year_df.dropna()  # Remove NaN rows (first year of each group)
world_energy_consumption_by_source_and_year_df
```

```python
world_energy_consumption_by_source_and_year_base = alt.Chart(world_energy_consumption_by_source_and_year_df).encode(
    x=alt.X("Year:O", title="Year"),
    y=alt.Y("Energy Consumption:Q", title="Energy Consumption (TWh)"),
    color=alt.Color("Clean Energy Source:N", title="Source", legend=alt.Legend(orient="right", offset=100))
)

world_energy_consumption_by_source_and_year_lines = world_energy_consumption_by_source_and_year_base.mark_line().encode(
    tooltip=["Year", "Clean Energy Source", "Energy Consumption"]
)
world_energy_consumption_by_source_and_year_labels = world_energy_consumption_by_source_and_year_base.mark_text(align="left", dx=5).encode(
    text="Clean Energy Source:N"
).transform_filter(
    alt.datum.Year == 2024
)

(world_energy_consumption_by_source_and_year_lines + world_energy_consumption_by_source_and_year_labels).properties(
    title="Global Clean Energy Consumption Over Time (1965-2024)",
    width=800,
    height=400,
    padding={"left": 5, "top": 5, "right": 120, "bottom": 5}
)
```

Observations:
- Hydropower has been and steady increase over the years
- Nuclear reachs its peak arounf 2000 and staggering since then.
- Wind and Solar is exponentially increasing in the past two decades.

```python
change_of_world_energy_consumption_by_source_and_year_base = alt.Chart(world_energy_consumption_by_source_and_year_df).encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.Y('Percent Change:Q', title='Year-over-Year Growth (%)', scale=alt.Scale(domain=[-20, 100], clamp=True)),
    color=alt.Color('Clean Energy Source:N', legend=alt.Legend(orient="right", offset=100))
)

change_of_world_energy_consumption_by_source_and_year_lines = change_of_world_energy_consumption_by_source_and_year_base.mark_line().encode(
    tooltip=['Year', 'Clean Energy Source', 'Percent Change']
)

change_of_world_energy_consumption_by_source_and_year_labels_right = change_of_world_energy_consumption_by_source_and_year_base.mark_text(align='left', dx=5).encode(
    text='Clean Energy Source:N'
).transform_filter(
    alt.datum.Year == 2024
)

(change_of_world_energy_consumption_by_source_and_year_lines + change_of_world_energy_consumption_by_source_and_year_labels_right).properties(
    title='Global Clean Energy Growth Rates (%)',
    width=800,
    height=400,
    padding={'left': 5, 'top': 5, 'right': 120, 'bottom': 5}
)
```
