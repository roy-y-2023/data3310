```python
import pandas as pd
```

```python
df = pd.read_csv("sunshine.csv")
df["monthnum"] = df["monthnum"].apply(lambda x: x + 1)
df
```

```python
import altair as alt
```

```python
chart = alt.Chart(df).mark_line().encode(
    x="monthnum",
    y="sunshine",
    color="city:N" # N denotes nominal/categorical data
)

chart
```

```python
# Calculate the last point for each city to place labels
last_points = df.loc[df.groupby('city')['monthnum'].idxmax()]

# Base line chart
line_chart = alt.Chart(df).mark_line().encode(
    x=alt.X("monthnum", title="Month"),
    y=alt.Y("sunshine", title="Sunshine Hours"),
    color=alt.Color("city:N", legend=None),
)

# Text labels at the end of each line
text_chart = alt.Chart(last_points).mark_text(
    align="left",
    baseline="middle",
    dx=5  # Offset to the right of the point
).encode(
    x="monthnum",
    y="sunshine",
    text="city:N",
    color=alt.Color("city:N", legend=None),
)

# Layer the charts
chart = (line_chart + text_chart).properties(title="Sunshine Hours by City")
chart
```
